```python
import googletrans
from googletrans import Translator
import pandas as pd
from tqdm import tqdm
```

```python
from googletrans import Translator
translator = Translator(service_urls=[
      'translate.google.com',
      'translate.google.co.uk',
      'translate.google.nl'
    ])
```

```python
df = pd.read_csv('dataentries.csv')
```

```python
df.head()
```

```python
if "Translated" in df.columns:
    pass
else:
    df.insert(4, "Translated", None)
```

```python
df.head()
```

```python
result = list()
for i, item in tqdm(enumerate(df['caption'])):

    if df.at[i,'is_translated'] == 0:
        try:
            translations = translator.translate([item], dest='fa')
            for translation in translations:
                    try:
                        df.at[i,'is_translated'] = 1 
                        df.at[i,'Translated'] = translation.text
                    except:
                        pass
        except:
            print("There is not enough volume.")
            break
            
print('Done!')
```
