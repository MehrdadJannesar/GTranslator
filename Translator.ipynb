```python
import googletrans
from googletrans import Translator
import pandas as pd
```

```python
from googletrans import Translator
translator = Translator(service_urls=[
      'translate.google.com',
      'translate.google.co.uk',
      'translate.google.nl'
    ])
```

```python
df = pd.read_csv('dataentries.csv')
```

```python
df.head()
```

```python
df_clear = df.drop(['origin_id', 'type', 'is_translated', 'created_at', 'updated_at'], axis=1)
```

```python
df_clear = df_clear[:100]
```

```python
df_clear
```

```python
result = list()
for item in df_clear['caption'][:100]:
    translations = translator.translate([item], dest='fa')
    for translation in translations:
        result.append(translation.text)
print('Done!')
```

```python
df_clear.insert(3, "Translate", result) 
```
